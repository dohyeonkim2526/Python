```python
from sklearn.decomposition import PCA
from sklearn import datasets
import numpy as np

X,y=datasets.make_regression(n_samples=200, n_features=10, n_informative=3, effective_rank=8)
```

```python
#pca 기능 확인
pca=PCA(n_components=None) #생성할 수 있는 모든 component 생성
pca.fit(X)

X.mean(0) #각 column별 평균
pca.mean_ #위와 같은 결과를 가지고 있음

comp=pca.components_ #생성된 datamatrix는 W^T를 의미함
exp_var=pca.explained_variance_ #각 component의 eigen value를 가지고 있음
```

```python
X0=X-pca.mean_ #각column의 평균이 0이 되게 만듦(작은 오차가 발생할 수 밖에 없음)

cov_mat=np.cov(X0.T)

np.diag(cov_mat).sum() #각 변수들의 variance 출력
exp_var.sum() #위와 같은 결과를 가지고 있음(오차 발생)
```

```python
t=np.dot(X0,comp[0]) #사영된 값 200개가 출력됨
```

```python
import matplotlib.pyplot as plt

cum_exp_var_ratio=np.cumsum(pca.explained_variance_ratio_)

plt.plot(range(1,11), cum_exp_var_ratio)
```
